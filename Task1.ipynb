```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df.drop(columns=['Country Code','Indicator Name','Indicator Code'],inplace=True)
```

```python
country_population = df.groupby('Country Name').mean().reset_index()
country_population = country_population.sort_values(by='2022', ascending=False)
top_countries = country_population.head(10)
print(top_countries)
plt.xticks(rotation=90)
plt.bar(top_countries['Country Name'], top_countries['2022'], color="#f25ed0")
plt.title("Top 10 Countries with Highest Population in 2022")
plt.xlabel("Country Name")
plt.ylabel("Population")
plt.show()
```
